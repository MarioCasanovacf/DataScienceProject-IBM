## Define Graphing Function


In this section, we define the function `make_graph`. You don't have to know how the function works, you should only care about the inputs. It takes a dataframe with stock data (dataframe must contain Date and Close columns), a dataframe with revenue data (dataframe must contain Date and Revenue columns), and the name of the stock.

Student comment: For a better stock representation, I decided to make a graph with the candlestick form and an alternative plot for the revenue that shows every trimestral report change.


````python
def make_graph(stock_data, revenue_data, stock):
    stock_data_specific = stock_data[stock_data.Date <= '2021-06-14']
    revenue_data_specific = revenue_data[revenue_data.Date <= '2021-04-30']

    # We will need to transform the column 'Date' a datetime
    revenue_data_specific['Date'] = pd.to_datetime(revenue_data_specific['Date'], infer_datetime_format=True)

    fig = make_subplots(rows=2, cols=1, subplot_titles=("Stock Prices", "Revenue"), shared_xaxes=True, vertical_spacing=.3)
    fig.add_trace(go.Candlestick(x=stock_data_specific.Date,
                                 open=stock_data_specific.Open,
                                 high=stock_data_specific.High,
                                 low=stock_data_specific.Low,
                                 close=stock_data_specific.Close,
                                 increasing_line_color='green',
                                 decreasing_line_color='red',
                                 increasing_line_width=1,
                                 decreasing_line_width=1,
                                 name="Share Price"), row=1, col=1)

    for year in revenue_data_specific.Date.dt.year.unique():
        year_data = revenue_data_specific[revenue_data_specific.Date.dt.year == year]
        fig.add_trace(go.Scatter(x=pd.to_datetime(year_data.Date, infer_datetime_format=True),
                                 y=year_data.Revenue.astype("float"),
                                 name=str(year),
                                 mode='lines',
                                 line=dict(dash="dot", shape='spline'),
                                 connectgaps=True,
                                 ), row=2, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=1, rangeslider=dict(visible=False))
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price", row=1, col=1)
    fig.update_yaxes(title_text="Revenue", tickprefix="$", row=2, col=1)
    fig.update_layout(height=900, hovermode='x unified')
    fig.show()
````
